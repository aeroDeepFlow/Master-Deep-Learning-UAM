```python
import pandas as pd
import numpy as np
```

```python
def medias_mayor_que(dataframe, limite):
    """
    Función que devuelve los nombres de las columnas cuya media es estrictamente superior al límite indicado
    
    PUNTUACIÓN: 2 PUNTOS
    
    Recibe:
        dataframe: Tabla de datos original
        límite: Valor numérico para su comparación
               
    Devuelve:
        lista de strings
        
    """
##########################################
#
#
#       CÓDIGO A IMPLEMENTAR
#
#
##########################################

    return dataframe.columns[dataframe.mean() > limite].to_list()
```

```python
def ordenar_columnas_valor_maximo(dataframe):
    """
    Devuelve un dataframe con las mismas columnas que el de entrada, pero con sus columnas dispuestas
    según el valor máximo que contenga cada columna, siendo la primera la mayor
    
    PUNTUACIÓN: 2 PUNTOS
    
    Recibe:
        dataframe: Tabla original
        
               
    Devuelve:
        dataframe
    """
##########################################
#
#
#       CÓDIGO A IMPLEMENTAR
#
#
##########################################

    return dataframe[dataframe.max().sort_values(ascending=False).index]
```

```python
def ultima_fila(dataframe, func, etiqueta):
    """
    Dado un dataframe, devuelve el mismo dataframe añadiendo una última fila aplicando una función a cada una
    de las columnas. El índice, ha de ser convertido a string para poder aplicar la etiqueta pasada como identificador
    de la nueva fila.
    
    PUNTUACIÓN: 2 PUNTOS
    
    Recibe:
        dataframe: El dataframe orignial.
        
        func: Función a aplicar a cada una de las columnas del dataframe original.
       
        etiqueta: Identificador de la nueva fila.
               
    Devuelve:
        Un dataframe
    """
##########################################
#
#
#       CÓDIGO A IMPLEMENTAR
#
#
##########################################

    return dataframe.append(dataframe.apply(func).to_frame(etiqueta).T)
```

```python
def incrementos(dataframe):
    """
    Dado un dataframe, devuelve un dataframe con las mismas columnas pero con la diferencia entre el valor de una celda
    y la anterior. Ejemplo:
    
    Fecha------|----A----       Fecha------|----A----     
    31/12/2018 |     500   ->   31/12/2018 |     300   
    30/12/2018 |     200        30/12/2018 |      NA
    
    
    PUNTUACIÓN: 2 PUNTOS
    
    Recibe:
        Dataframe: Dataframe orignal
               
    Devuelve:
        Dataframe
    """
##########################################
#
#
#       CÓDIGO A IMPLEMENTAR
#
#
##########################################

    return dataframe - dataframe.shift(periods=-1)
```

```python
def no_nulos(dataframe):
    """
    Devuelve un dataframe con una fila extra indicando el número de no nulos de cada columna
    
    PUNTUACIÓN: 2 PUNTOS
    
    Recibe: 
        
        Dataframe: dataframe original
        
               
    Devuelve:
        Dataframe
        
    """ 
    
##########################################
#
#
#       CÓDIGO A IMPLEMENTAR
#
#
##########################################

    return len(dataframe) - dataframe.isnull().sum().to_frame("No Nulos").T
```
