```python
olympics = pd.read_csv("olympics.csv", header=1)

# Diccionario para renombrar las columnas
cols_renamed = {
    "Unnamed: 0": "pais",
    "? Summer": "verano_participaciones",
    "01 !": "verano_oro",
    "02 !": "verano_plata",
    "03 !": "verano_bronce",
    "Total": "verano_total",
    "? Winter": "invierno_participaciones",
    "01 !.1": "invierno_oro",
    "02 !.1": "invierno_plata",
    "03 !.1": "invierno_bronce",
    "Total.1": "invierno_total",
    "? Games": "combinados_participaciones",
    "01 !.2": "combinados_oro",
    "02 !.2": "combinados_plata",
    "03 !.2": "combinados_bronce",
    "Combined total": "combinados_total"
}

olympics.rename(columns=cols_renamed, inplace=True)

# Establecer país como índice del dataframe
olympics.set_index("pais", drop=True, inplace=True)

# Eliminar la fila de totales
olympics = olympics[olympics.index != "Totals"]

# Crear las columnas como multiíndice
olympics.columns = pd.MultiIndex.from_tuples(
    [col.split("_") for col in olympics.columns])

# Eliminar las abreviaciones en los nombres de los países
olympics.index = olympics.index.str.split("(").str[0]
olympics.index = olympics.index.str.rstrip()
```
