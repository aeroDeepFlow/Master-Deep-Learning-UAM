```python
test_file = "Practica5_Evaluable.ipynb"
%run $test_file

title = "World historical and predicted populations (in millions)"
url = "https://en.wikipedia.org/wiki/World_population"
```

```python
%run $test_file

def test_get_world_population(title, url):
    
    
    expected = {
        ("World", "1600"): 660,
        ("Europe", "1850"): 276,
        ("Oceania", "1999"): 30,
    }
      
    df = get_world_population(title, url)
    print(df.columns)

    try:
        for key, population in expected.items():
            continent, year = key

            received = df.loc[continent,year]
            assert received == population 
            
        print("get_world_population test passed!")   
        return df
    
    except AssertionError:
        print("test failed at %s %s. Expected %i, received: %i" % (continent, year, population, received))
        return None
    
    
df = test_get_world_population(title, url)
```
